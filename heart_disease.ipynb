```python
import pandas as pd 
import numpy as np
import matplotlib as mpl
import matplotlib.pyplot as  plt 
```

```python
%matplotlib inline
```

```python
my_df = pd.read_csv('heart_disease_dataset.csv')
```

divide smokers by age

```python
dataf = {
        ' ':['Smoking','NoSmoking'],
        '30 to 40':[0,0],
        '30 to 40 (%)':[0,0],
        '41 to 50':[0,0],
        '41 to 50 (%)':[0,0],
        '51 to 60':[0,0],
        '51 to 60 (%)':[0,0],
        '61 to 70':[0,0],
        '61 to 70 (%)':[0,0]
        }
my_smoker = pd.DataFrame(dataf)
```

```python
my_smoker['30 to 40'][0] = my_df.value_counts((my_df['age']>=30) & (my_df['age']<41) & (my_df['currentSmoker']==1))[1] 
my_smoker['30 to 40'][1] = my_df.value_counts((my_df['age']>=30) & (my_df['age']<41) & (my_df['currentSmoker']==0))[1]
my_smoker['30 to 40 (%)'][0] = round((my_smoker['30 to 40'][0] / (my_smoker['30 to 40'][1] + my_smoker['30 to 40'][0]))*100,2 )
my_smoker['30 to 40 (%)'][1] = 100 - my_smoker['30 to 40 (%)'][0]

my_smoker['41 to 50'][0] = my_df.value_counts((my_df['age']>40) & (my_df['age']<51) & (my_df['currentSmoker']==1))[1]
my_smoker['41 to 50'][1] = my_df.value_counts((my_df['age']>40) & (my_df['age']<51) & (my_df['currentSmoker']==0))[1]
my_smoker['41 to 50 (%)'][0] = round((my_smoker['41 to 50'][0] / (my_smoker['41 to 50'][1] + my_smoker['41 to 50'][0]))*100,2 )
my_smoker['41 to 50 (%)'][1] = 100 - my_smoker['41 to 50 (%)'][0]

my_smoker['51 to 60'][0] = my_df.value_counts((my_df['age']>50) & (my_df['age']<61) & (my_df['currentSmoker']==1))[1]
my_smoker['51 to 60'][1] = my_df.value_counts((my_df['age']>50) & (my_df['age']<61) & (my_df['currentSmoker']==0))[1]
my_smoker['51 to 60 (%)'][0] = round((my_smoker['51 to 60'][0] / (my_smoker['51 to 60'][1] + my_smoker['51 to 60'][0]))*100,2 )
my_smoker['51 to 60 (%)'][1] = 100 - my_smoker['51 to 60 (%)'][0]

my_smoker['61 to 70'][0] = my_df.value_counts((my_df['age']>60) & (my_df['age']<71) & (my_df['currentSmoker']==1))[1]
my_smoker['61 to 70'][1] = my_df.value_counts((my_df['age']>60) & (my_df['age']<71) & (my_df['currentSmoker']==0))[1]
my_smoker['61 to 70 (%)'][0] = round((my_smoker['61 to 70'][0] / (my_smoker['61 to 70'][1] + my_smoker['61 to 70'][0]))*100,2 )
my_smoker['61 to 70 (%)'][1] = 100 - my_smoker['61 to 70 (%)'][0]
```

```python
my_smoker
```

```python
df_smoking = [my_smoker['30 to 40 (%)'][0],my_smoker['41 to 50 (%)'][0], my_smoker['51 to 60 (%)'][0], my_smoker['61 to 70 (%)'][0]]
df_Nosmoking = [my_smoker['30 to 40 (%)'][1],my_smoker['41 to 50 (%)'][1], my_smoker['51 to 60 (%)'][1], my_smoker['61 to 70 (%)'][1]]
age = ['30 to 40', '41 to 50', '51 to 60', '61 to 70']
df_smoking
```

```python
plt.plot(age,df_smoking, color='red', label ='Smoking')
plt.grid()
plt.xlabel('age')
plt.ylabel('(%)')

plt.plot(age,df_Nosmoking, color='Blue', label = 'Nosmoking')
plt.legend(loc = 'best')

plt.show()
```

```python
dtf = {
    'age':['30 to 40', '41 to 50', '51 to 60', '61 to 70'],
    'cigsPerDay':[0,0,0,0]
}
CPD = pd.DataFrame(dtf)
```

```python

CPD['cigsPerDay'][0]=  round(my_df[(my_df['age'] >30) & (my_df['age']<41)].mean()['cigsPerDay'],2)
CPD['cigsPerDay'][1] = round(my_df[(my_df['age'] >40) & (my_df['age']<51)].mean()['cigsPerDay'],2)
CPD['cigsPerDay'][2] = round(my_df[(my_df['age'] >50) & (my_df['age']<61)].mean()['cigsPerDay'],2)
CPD['cigsPerDay'][3] = round(my_df[(my_df['age'] >60) & (my_df['age']<71)].mean()['cigsPerDay'],2)
CPD
```

```python

cigarette = CPD['age'].unique()
CP = CPD['cigsPerDay']
plt.bar(x=cigarette, height=CP)
plt.xticks(cigarette, size = 15)
plt.xlabel('age',size = 15)
plt.ylabel('cigsPerDay')
plt.show()
```
